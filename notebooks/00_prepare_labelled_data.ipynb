```python
import pandas as pd 
from pathlib import Path 
import os
os.chdir(r"C:\Users\josep\Desktop\repos\com774-mlops-code-quality")



INPUT_PATH = Path("data/processed/sonarqube_cleaned_cw1.csv")
OUTPUT_PATH = Path("data/processed/sonarqube_cleaned_cw2.csv")

df = pd.read_csv(INPUT_PATH)

# Inspect the problematic_ratio column
print(df["problematic_ratio"].describe())

THRESHOLD = 0.03

df["high_risk"] = (df["problematic_ratio"] > THRESHOLD).astype(int)

print(df["high_risk"].value_counts(normalize=True))

OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved labelled dataset to: {OUTPUT_PATH}")
```
